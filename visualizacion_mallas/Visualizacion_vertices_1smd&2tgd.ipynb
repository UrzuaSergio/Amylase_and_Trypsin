```python
import numpy
from matplotlib import pyplot as plt
```

### Visualización de Amilasa junto con sensor: Verificación Tamaño

```python
file_1smd_rot = open("1smd_d08_rot0-tilt90-h2.vert", "r")
X_1smd_rot = numpy.loadtxt(file_1smd_rot)
file_sensor_50 = open("sensor_50x50x10_d02.vert", "r")
X_sensor_50 = numpy.loadtxt(file_sensor_50)
file_sensor_100 = open("sensor_100x100x10_d02.vert", "r")
X_sensor_100 = numpy.loadtxt(file_sensor_100)
```

```python
print("con sensor de 100x100x10")
print("vista x-z")
plt.plot(X_1smd_rot[:,0],X_1smd_rot[:,2])
plt.plot(X_sensor_50[:,0],X_sensor_50[:,2])
plt.show()
print("vista x-y")
plt.plot(X_1smd_rot[:,0],X_1smd_rot[:,1])
plt.plot(X_sensor_50[:,0],X_sensor_50[:,1])
plt.show()

print("con sensor de 100x100x10")
print("vista x-z")
plt.plot(X_1smd_rot[:,0],X_1smd_rot[:,2])
plt.plot(X_sensor_100[:,0],X_sensor_100[:,2])
plt.show()
print("vista x-y")
plt.plot(X_1smd_rot[:,0],X_1smd_rot[:,1])
plt.plot(X_sensor_100[:,0],X_sensor_100[:,1])
plt.show()
```
